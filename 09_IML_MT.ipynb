## **UNIVARIATE** **ANALYSIS**

## **JOB** **SATISFACTION** 

```python
# univariate frequency distribution

tab = pd.crosstab(attrition.JobSatisfaction, columns='Frequency')
tab
```

```python
#univariate table of proportion 

n = len(attrition)
tab/n
```

Nearly 20% of the company's employees under our observation are not satisfied with their job. 

```python
attrition.groupby('JobSatisfaction').Attrition.mean()
```
